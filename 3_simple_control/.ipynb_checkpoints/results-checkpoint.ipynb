```python
import matplotlib.pyplot as plt
import pandas as pd

colors = ['b', 'g', 'r', 'c']
```

```python
data = pd.read_csv("csv/pid_1_0_0.csv")
column_names = data.columns
fig, axes = plt.subplots(len(column_names), figsize=(10, 6))
for i, column_name in enumerate(column_names):
    axes[i].plot(data[column_name][:500], color=colors[i])
    axes[i].set_title(column_name)
plt.xlabel("Time")
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv("csv/pid_5_0_0.csv")
column_names = data.columns
fig, axes = plt.subplots(len(column_names), figsize=(10, 6))
for i, column_name in enumerate(column_names):
    axes[i].plot(data[column_name][:500], color=colors[i])
    axes[i].set_title(column_name)
plt.xlabel("Time")
plt.tight_layout()
plt.show()
```

1) we can observer bigger overshoots
2) but rms converges better
